```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
```

```python
data = {
    "Jam Belajar" : [2,3,4,5,6],
    "Skor Ujian" : [50,60,65,70,80]
}
df = pd.DataFrame(data)
pearson_corr, p_value = pearsonr(df["Jam Belajar"], df["Skor Ujian"])
spearman_corr, _ = spearmanr(df["Jam Belajar"], df["Skor Ujian"])

sns.lmplot(x="Jam Belajar", y="Skor Ujian", data=df, ci=None)
plt.title(f"Korelasi Pearson : {pearson_corr:.2f}, Korelasi Spearman : {spearman_corr:.2f}")
plt.xlabel("Jam Belajar")
plt.ylabel("Skor Ujian")
plt.show()
```
