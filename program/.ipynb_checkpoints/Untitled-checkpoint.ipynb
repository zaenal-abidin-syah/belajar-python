```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = {
    "Pendapatan" :[15, 16, 17, 25, 26,27, 35,36,37],
    "Pengeluaran" : [80, 85,90,50,55,60, 20,25,30]
}
df = pd.DataFrame(data)
```

```python
kmean = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmean.fit_predict(df)
```

```python
plt.scatter(df["Pendapatan"], df["Pengeluaran"], c=df["Cluster"], cmap="viridis", s=100)
plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], color="red", marker="x", label="Centroid")
plt.xlabel("Pendapatan")
plt.ylabel("Pengeluaran")
plt.legend()
plt.title("Klasterisasi Pelanggan")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
luas_tanah = np.array([50,60,70,80,90]).reshape(-1,1)
harga_rumah = np.array([150, 200,250,300,350])
```

```python
model = LinearRegression()
model.fit(luas_tanah, harga_rumah)
```

```python
prediksi = model.predict(luas_tanah)
mse = mean_squared_error(harga_rumah, prediksi)
```

```python
print(f"Mean Squared Error : {mse:.2f}")
print(f"Prediksi harga untuk luas tanah 75 m2: {model.predict([[75]])[0]}")
```

```python
plt.scatter(luas_tanah, harga_rumah, color="blue", label='Data Aktual')
plt.plot(luas_tanah, prediksi, color='red', label='Model Prediksi')
plt.xlabel('Luas tanah (m^2)')
plt.ylabel('Harga rumah (juta Rupiah)')
plt.legend()
plt.title('Regresi Linier: Harga Rumah vs Luas Tanah')
plt.show()
```

```python
from scipy.optimize import linprog
```

```python
profit = [-20,-30]
A = [[1,2],
    [3,1]]
B = [40,60]
x_bounds = (0, None)
bounds = [x_bounds, x_bounds]
result = linprog(profit, A_ub=A, b_ub=B, bounds=bounds, method="highs")


print(f"Nilai variabel x: {result.x}")
print(f"Profit maksimal: {-result.fun}")
```
